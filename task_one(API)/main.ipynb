1. Import Required Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

2. Load the Dataset

Use pandas to load datasets 

```python
# Load the datasets
api_data = pd.read_csv('./T1_data/API.csv')

# Preview the first few rows to understand the structure
api_data.head()
```

3. Data Cleaning and Filtering

Identify and drop unnecessary or NaN rows and columns.

Extract relevant columns such as Country Code and Population Data.

```python
# Drop rows with any NaN values in the population dataset
api_data_cleaned = api_data.dropna()
```

```python
# Select the second column and columns from the fifth to the end
# Note: Indexing starts at 0, so 1 is the second column, and 4 is the fifth column
selected_columns = api_data_cleaned.iloc[:, [1] + list(range(4, api_data_cleaned.shape[1]))]

# Get the names of the selected columns
column_names = selected_columns.columns.tolist()

# Print the selected DataFrame and its column names
print(selected_columns.head())
print("Selected column names:", column_names)
```

```python
# Select relevant columns (e.g., 'Country Code' and 'Years')
population_data = api_data_cleaned[column_names]
print(population_data.head())
```

```python
column_names[1:]
```

4. Create a Bar Chart and Histogram

```python
# Sort the population data in descending order and select the top 10
top_10 = population_data.sort_values(by='2019', ascending=False).head(10)

# Create a bar chart
plt.figure(figsize=(12, 6))
sns.barplot(x='2019', y='Country Code', data=top_10, palette='viridis')
plt.xlabel('Population (2019)')
plt.ylabel('Country')
plt.title('Top 10 Countries by Population in 2019')
plt.show()
```

```python
# Create a histogram to visualize the distribution of population
plt.figure(figsize=(10, 6))
sns.histplot(population_data['2019'], bins=30, kde=True)
plt.xlabel('Population (2019)')
plt.title('Distribution of Population in 2019')
plt.show()
```

# Plot the population from 1960 to 2023 using line plot

Step 1: Reshape the Data
Use pd.melt() to transform the data from wide to long format, which makes it easier to work with time-series data.

```python
#Reshape the data to long format
population_long = pd.melt(
    api_data_cleaned,
    id_vars=['Country Name', 'Country Code'],
    var_name='Year',
    value_name='Population'
)
population_long
```

Step 2: Convert Data Types
Ensure that the 'Year' and 'Population' columns are numeric to avoid errors and enable proper sorting and plotting.

```python
# Convert the 'Year' column to numeric to allow sorting and plotting
population_long['Year'] = pd.to_numeric(population_long['Year'], errors='coerce')

# Convert 'Population' to numeric, setting errors='coerce' to handle any non-numeric entries
population_long['Population'] = pd.to_numeric(population_long['Population'], errors='coerce')
population_long
```

Step 3: Clean the Data
Remove rows with missing population values to avoid errors when plotting or filtering.

```python
# Drop rows with NaN values in 'Population' (if any)
population_long = population_long.dropna(subset=['Population'])
population_long.head()
```

Step 4: Filter Top 5 Countries by Population in 2023 Identify the top 5 countries with the largest population for the year 2023.

```python
# Filter the top 5 countries by population in 2023
top_countries = (
    population_long[population_long['Year'] == 2023]
    .nlargest(5, 'Population')['Country Code']
)
top_countries
```

Step 5: Filter Data for Top 5 Countries Across All Years
Extract the data for the identified top 5 countries across all years.

```python
# Filter the data to include only the top 5 countries across all years
filtered_data = population_long[population_long['Country Code'].isin(top_countries)]
filtered_data
```

Step 6: Plot the Population Trend for Top 5 Countries
Create a line plot to visualize population trends for these countries over the years.

```python
# Plot the population trend for the top 5 countries
plt.figure(figsize=(14, 8))
sns.lineplot(data=filtered_data, x='Year', y='Population', hue='Country Code', marker='o')

plt.title('Population Trend (All Years) for Top 5 Countries by 2023 Population')
plt.xlabel('Year')
plt.ylabel('Population')
plt.legend(title='Country', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

# Next Steps
Letâ€™s now strictly filter for valid ISO country codes that correspond to known countries. We will exclude all regional, income-based, or program-based classifications by:

- Keeping only codes that match real-world ISO country codes.  
- For simplicity, focus on known countries such as China (CHN), India (IND), USA (USA), etc.

```python
# Step 1: Filter only valid three-letter ISO codes for countries
valid_countries = population_long[
    population_long['Country Code'].isin([
        'CHN', 'IND', 'USA', 'IDN', 'PAK', 'BRA', 'NGA', 'BGD', 'RUS', 'JPN'
    ])
]

# Check the filtered data to ensure only actual countries are included
valid_countries.head()


```

```python
# Step 2: Get the top 5 most populous countries in 2023
top_5_countries_2023 = (
    valid_countries[valid_countries['Year'] == 2023]
    .nlargest(5, 'Population')
)

# Print the Top 5 Countries in 2023
print("Top 5 Countries by Population in 2023:")
print(top_5_countries_2023[['Country Name', 'Country Code', 'Population']])
```

```python
# Step 3: Filter the data for the identified top 5 countries across all years
filtered_top_5 = valid_countries[
    valid_countries['Country Code'].isin(top_5_countries_2023['Country Code'])
]

# Preview the filtered data
filtered_top_5.head()
```

```python
# Step 4: Create a line plot for population trends of the top 5 countries
plt.figure(figsize=(14, 8))
sns.lineplot(data=filtered_top_5, x='Year', y='Population', hue='Country Name', marker='o')

# Add labels and title
plt.title('Population Trends (1960-2023) for Top 5 Countries by 2023 Population')
plt.xlabel('Year')
plt.ylabel('Population')
plt.legend(title='Country', bbox_to_anchor=(1.05, 1), loc='upper left')

# Show the plot
plt.show()
```
